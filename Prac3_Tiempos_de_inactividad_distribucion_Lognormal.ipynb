```python
# Alumna: Romero Jiménez Sofía Cristal

import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as stats
```

```python
# Parámetros de la distribución lognormal
mu = 1.0   # Media de la variable logarítmica
sigma = 0.5  # Desviación estándar de la variable logarítmica
n = 100  # Número de variables aleatorias a generar
```

```python
# Generar n números aleatorios U ~ Uniform(0, 1)
U = np.random.uniform(0, 1, n)
```

```python
# Transformada inversa para generar tiempos de inactividad lognormales
Z = mu + sigma * stats.norm.ppf(U)
T = np.exp(Z)  # Tiempos generados
```

```python
# Mostrar los resultados
print("Tiempos de inactividad generados:", T)
```

```python
# Diseño de la grafica
color_barras = '#B2FF59'
color_borde = '#212121'
color_fondo = '#F8BBD0'
color_texto = '#4A148C'
color_cuadricula = '#FFD600'
```

```python
# Graficar histograma de los tiempos de inactividad generados
plt.figure(figsize=(8, 6))
plt.hist(T, bins=20, color=color_barras, edgecolor=color_borde, alpha=0.85, density=True)

# Graficar la función teórica de densidad de la distribución lognormal
x = np.linspace(min(T), max(T), 100)
pdf = stats.lognorm.pdf(x, sigma, scale=np.exp(mu))
plt.plot(x, pdf, color='#FF4081', linewidth=2, label='Distribución lognormal teórica')  # Rosa intenso

plt.title("Histograma de los tiempos de inactividad (Distribución Lognormal)", color=color_texto, fontsize=14, fontweight='bold')
plt.xlabel("Tiempo de inactividad", color=color_texto, fontsize=12)
plt.ylabel("Densidad de probabilidad", color=color_texto, fontsize=12)
plt.gca().set_facecolor(color_fondo)  # Fondo del gráfico
plt.grid(axis='y', linestyle='--', color=color_cuadricula, alpha=0.6)
plt.legend()
plt.show()
```
