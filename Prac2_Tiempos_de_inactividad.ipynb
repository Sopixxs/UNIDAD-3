```python
# Parámetro de la distribución exponencial (tasa de inactividad)
lambda_inactividad = 1 / 10  # Tiempo promedio de inactividad = 10 minutos
```

```python
# Función para generar un tiempo de inactividad
def generar_tiempo_inactividad(lambda_inactividad):
    u = np.random.uniform(0, 1)  # Generar número aleatorio uniforme
    # Aplicar transformada inversa
    tiempo_inactividad = -1 / lambda_inactividad * np.log(1 - u)
    return tiempo_inactividad
```

```python
# Generar 100 tiempos de inactividad
tiempos_inactividad = [generar_tiempo_inactividad(lambda_inactividad)
                       for _ in range(100)]
```

```python
# Mostrar los tiempos de inactividad generados
print("Tiempos de inactividad generados: ",
      tiempos_inactividad)
```

```python
# Diseño de la grafica
color_barras = '#FF80AB'
color_borde = '#D81B60'
color_fondo = '#FCE4EC'
color_texto = '#283593'
color_cuadricula = '#90CAF9'
```

```python
# Graficar los tiempos de inactividad simulados
plt.figure(figsize=(8, 6))
plt.hist(tiempos_inactividad, bins=10, color=color_barras, edgecolor=color_borde, alpha=0.85)
plt.title('Distribución de Tiempos de Inactividad Simulados (Exponencial)', color=color_texto, fontsize=14, fontweight='bold')
plt.xlabel('Tiempo de inactividad (minutos)', color=color_texto, fontsize=12)
plt.ylabel('Frecuencia', color=color_texto, fontsize=12)
plt.gca().set_facecolor(color_fondo)  # Fondo del gráfico
plt.grid(axis='y', linestyle='--', color=color_cuadricula, alpha=0.6)
plt.show()
```
