```python
# Parámetro de distribución exponencial
lambda_servicio = 1 / 2  # Tiempo promedio de servicio = 2 minutos
```

```python
# Función para generar 1 tiempo de servicio
def generar_tiempo_servicio(lambda_servicio):
  u = np.random.uniform(0, 1) # Genera un número aleatorio uniforme
  tiempo_servicio = -1 / lambda_servicio * np.log(1 - u) # Aplica la transformada inversa
  return tiempo_servicio
```

```python
# Genera 100 tiempos de servicio
tiempos_servicio = [generar_tiempo_servicio (lambda_servicio) for __ in range(100)]
```

```python
# Muestra los tiempos de servicio
print("Tiempos de servicio generados:", tiempos_servicio)
```

```python
# Diseño de la grafica
color_barras = '#E91E63'
color_borde = '#880E4F'
color_fondo = '#FCE4EC'
color_texto = '#4A148C'
```

```python
# Grafica los tiempos de servicio simulados
plt.figure(figsize=(8, 6))
plt.hist(tiempos_servicio, bins=10, color=color_barras, edgecolor=color_borde, alpha=0.85)
plt.title('Distribución de Tiempos de Servicio Simulados (Exponencial)', color=color_texto, fontsize=14, fontweight='bold')
plt.xlabel('Tiempo de servicio (minutos)', color=color_texto, fontsize=12)
plt.ylabel('Frecuencia', color=color_texto, fontsize=12)
plt.gca().set_facecolor(color_fondo)
plt.grid(axis='y', linestyle='--', alpha=0.6)
plt.show()
```
